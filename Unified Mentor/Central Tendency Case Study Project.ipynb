## view the distribution of the various features in the dataset and calculated their  central tensencies

### now we will look  at the distribution of the various  features in dataset , we will also calculate appx measures of central tendency for these features 

```python
df.columns
```

```python
#create a histogram of the fixed acidity feature

plt.figure(figsize=(10,6))

sns.histplot(data = df ,x='fixed acidity',color='orange',edgecolor='linen',alpha=0.5,bins =5,kde=True)
plt.title('Histogram of fixed acidity')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.show()
```

```python
round(df['fixed acidity'].mean(),2)
```

```python
df['fixed acidity'].median()
```

```python
#create a histogram of the fixed acidity feature include mean and median values

plt.figure(figsize=(10,6))

sns.histplot(data = df ,x='fixed acidity',color='orange',edgecolor='linen',alpha=0.5,bins =5,kde=True)
plt.title('Histogram of fixed acidity')
plt.xlabel('fixed acidity')
plt.ylabel('count')
plt.vlines(df['fixed acidity'].mean(),ymin=0,ymax=4000,color='blue',label='Mean')
plt.vlines(df['fixed acidity'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

```python
df.columns
```

```python
# df['volatile acidity']
#create a histogram of the 'volatile acidity' feature include mean and median values

plt.figure(figsize=(10,6))

sns.histplot(data = df ,x='volatile acidity',color='green',edgecolor='linen',alpha=0.5,bins =5)
plt.title('Histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.show()
```

#### here volatile acidity is not distributed ,it is skewed a lilltle towaqrds the right

```python
plt.figure(figsize=(10,6))
sns.distplot(x=df['volatile acidity'])
plt.title('Distplot of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('density')
plt.show()
```

```python
#check the skeweness
df['volatile acidity'].skew()
```

```python
# df['volatile acidity']
#create a histogram of the 'volatile acidity' feature include mean and median values

plt.figure(figsize=(10,6))

sns.histplot(data = df ,x='volatile acidity',color='green',edgecolor='linen',alpha=0.5,bins =5)
plt.title('Histogram of volatile acidity')
plt.xlabel('volatile acidity')
plt.ylabel('count')
plt.vlines(df['volatile acidity'].mean(),ymin=0,ymax=4000,color='blue',label='Mean')
plt.vlines(df['volatile acidity'].median(),ymin=0,ymax=4000,color='red',label='Median')
plt.legend()
plt.show()
```

### create a new Pandas  Series called 'rep_acid' that contains with detail of the representative quality for the diffreent types of acids

```python
rep_acid = pd.Series(index=['fixed acidity','volatile acidity', 'citric acid','quality'],
                     data=[df['fixed acidity'].mean(),df['volatile acidity'].mean(),
                    df['citric acid'].mean(),df['quality'].value_counts().index[0]])
```

```python
rep_acid
```

## the representitve acid of the quality is as follows:
    
1. the mean value of the fixed acidity would be 6.85
2. the mean value of the voltile acid would be 0.27
3. the mean value of the citric acid would be 6.
4. the quality  would be 6.
