## visualize the distribution of ratings and compensation

```python
#create a scatterplot of the  'age' and 'Employeesalary ' features
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='age',y='salary',color='r',edgecolor='linen',alpha=0.5)
plt.title("Scatter plot of Employee age vs Employee salary")
plt.xlabel('EmployeeAge')
plt.ylabel('EmployeeSalary')
plt.show()
```

## Observation
we observe that in general, there isa realtionship btw age and Salary of employee
as a age increases, the salary also increases, though there some outliers present as well

```python
#create a scatterplot of the  'healthy_eating', and'active_lifestyle', features
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='healthy_eating',y='active_lifestyle',color='g',edgecolor='linen',alpha=0.5)
plt.title("Scatter plot of 'healthy_eating vs active_lifestyle")
plt.xlabel('healthy_eating')
plt.ylabel('active_lifestyle')
plt.show()
```

```python
#create a scatterplot of the  'healthy_eating', and'Salary', features
plt.figure(figsize=(10,6))
sns.scatterplot(data=df,x='healthy_eating',y='salary',color='y',edgecolor='linen',alpha=0.5)
plt.title("Scatter plot of 'healthy_eating vs salary")
plt.xlabel('healthy_eating')
plt.ylabel('salary')
plt.show()
```

```python
#create a countplot  of the group feature
plt.figure(figsize=(10,6))
sns.countplot(data=df,x='groups',edgecolor='linen',alpha=0.5)
plt.title("Count plot of groups")
plt.xlabel('groups')
plt.ylabel('count')
plt.show()
```

```python
#create histogram of salary
plt.figure(figsize=(5,5))
sns.histplot(data=df,x='salary',bins=10)
plt.show()
```

#we see  that  salaries  are not uniform , which is  not necessarily  a discrepancy
but we do observe that there are employee with salaries on either extremites of the histogram
 based on the histogram, we can see  that  the count of hieghest earnings employee is very less
